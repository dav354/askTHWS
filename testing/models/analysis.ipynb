```python
import pandas as pd
```

```python
data = pd.read_csv("/Users/lelange/Uni/Projektarbeit/rag/testing/test_results_scored.csv")
```

```python
# Mapping: Modellname → (Zeitspalte, Score-Spalte)
models = {
    "gemma_7b": ("answer_gemma7b_time", "score_gemma7b"),
    "orca_mini_7b": ("answer_orca-mini7b_time", "score_orca-mini7b"),
    "mistral_7b": ("answer_mistral_latest_time", "score_mistral_latest"),
    "phi4_latest": ("answer_phi4_latest_time", "score_phi4_latest"),
    "qwen25_14b": ("answer_qwen2_5_14b_time", "score_qwen2_5_14b"),
    "orca_mini_13b": ("answer_orca-mini_13b_time", "score_orca-mini_13b"),
    "gemma3_27b": ("answer_gemma3_27b_time", "score_gemma3_27b"),
    "deepseek_r1_32b": ("answer_deepseek_r1_32b_time", "score_deepseek_r1_32b"),
    "qwq_latest": ("answer_qwq_latest_time", "score_qwq_latest")
}

# Ergebnisse berechnen
results = []

for model_name, (time_col, score_col) in models.items():
    if time_col in data.columns and score_col in data.columns:
        time_series = pd.to_numeric(data[time_col], errors='coerce')
        score_series = pd.to_numeric(data[score_col], errors='coerce')
        avg_time = round(time_series.mean(skipna=True), 2)
        avg_score = round(score_series.mean(skipna=True), 4)
        percentage_07 = (score_series <= 0.7).sum() / len(data) * 100
        percentage_08 = (score_series >= 0.8).sum() / len(data) * 100
        percentage_09 = (score_series >= 0.9).sum() / len(data) * 100

        results.append({
            "model": model_name,
            "avg_time_to_answer": avg_time,
            "avg_score": avg_score,
            "worse_than_07": f"{percentage_07:.2f}%",
            "better_than_08": f"{percentage_08:.2f}%",
            "better_than_09": f"{percentage_09:.2f}%"
        })
    else:
        print(f"⚠️  Spalten fehlen für Modell '{model_name}': {time_col}, {score_col}")

# Ergebnis als DataFrame
summary_df = pd.DataFrame(results)
```
